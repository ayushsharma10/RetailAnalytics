#### Pages Visited

```python
pages_cols = ['Administrative', 'Informational', 'ProductRelated']
duration_cols = ['Administrative_Duration', 'Informational_Duration', 'ProductRelated_Duration']

for i in pages_cols:
    print('Boxplot and Distributon for', i, 'feature')
    f, ax = plt.subplots(1,2, figsize = (17,5))
    sns.boxplot(df[i], ax=ax[0])
    sns.distplot(df[i], ax=ax[1])
    plt.show()
```

Since majority of people have not viewed the page, let's find distribution for people who have viewed the pages

```python
for i in pages_cols:
    
    print('Boxplot and Distributon for', i, 'feature with atleast 1 page viewed')
    f, ax = plt.subplots(1,2, figsize = (17,5))
    sns.boxplot(df[df[i] != 0][i], ax=ax[0])
    sns.distplot(df[df[i] != 0][i], ax=ax[1])
    plt.show()
```

Despite taking away the zero distribution, our data is hevily positive skewed

#### Consumer behaviour to pages visits (average)

```python
df[pages_cols].groupby(df.Revenue).mean().T.plot.bar()
```

#### Duration on Websites

```python
f, ax = plt.subplots(1,3, figsize = (17,5))

sns.violinplot(y = df.Administrative_Duration, x = df.Revenue, ax = ax[0])
sns.violinplot(y = df.Informational_Duration, x = df.Revenue, ax = ax[1])
sns.violinplot(y = df.ProductRelated_Duration, x = df.Revenue, ax = ax[2])
```

```python
# Since data is heavily skewed, we try log transformation
f, ax = plt.subplots(1,3, figsize = (17,5))


sns.violinplot(y = df.Administrative_Duration.apply(lambda x: np.log(x)).replace(-np.inf,0), x = df.Revenue, ax = ax[0])
sns.violinplot(y = df.Informational_Duration.apply(lambda x: np.log(x)).replace(-np.inf,0), x = df.Revenue, ax = ax[1])
sns.violinplot(y = df.ProductRelated_Duration.apply(lambda x: np.log(x)).replace(-np.inf,0), x = df.Revenue, ax = ax[2])
```
